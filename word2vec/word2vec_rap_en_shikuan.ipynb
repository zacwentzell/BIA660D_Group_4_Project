```python
import numpy as np
import pandas as pd
import gensim
import nltk
from nltk.corpus import brown
```

```python
songs = pd.read_csv('C:/Users/alienware/Desktop/cleaned_data_rap_en.csv')
```

```python
Lyrics=list(songs['Lyrics'])
corpus=(Lyrics)
```

```python
rap_song=[]
Tokens=[]
for i in range(len(corpus)):
    g=corpus[i].split()
    rap_song.append(set(g))
    Tokens.append((g))
sentencesBrown = brown.sents()   
for i in range(len(sentencesBrown)):
    rap_song.append(set(sentencesBrown[i]))
```

```python
model = gensim.models.Word2Vec(rap_song, size=50,window=10,min_count=5)
```

```python
vocab=model.wv.vocab
vocab=(set(vocab))
```

```python
word2vecTokens=[]
i=0;
for g in Tokens:
    vc=[]
    for s in g:
        if (s in vocab):
            vc.append(model.wv[s])
    word2vecTokens.append(vc)
```

```python
g=[]
for i in range (len(word2vecTokens)):
    g.append(np.sum(word2vecTokens[i], axis=0)/len(word2vecTokens[i]))
```

```python
columns=[]
index=[]
for k in range(0,80000):
    index.append(k)
for i in range(1,51):
    columns.append("w2v_"+str(i))
df_ = pd.DataFrame(columns=columns)
df_ = df_.fillna(0)
```

```python
for x in range (len(g)):
    try:
        if(type(g[x])==float):
            g[x]=[0]*50
        g[x]=g[x].tolist()
    except:
        print(x)
```

```python
for x in range (len(g)):
    try:
        if(type(g[x])==float):
            g[x]=[0]*50
    except:
        print(x)
dar=pd.DataFrame(g,columns=columns)
dar
```

```python
dar.to_csv("word2vec_rap_en.csv",index=False)
```
