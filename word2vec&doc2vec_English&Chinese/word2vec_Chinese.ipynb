```python
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords  
```

```python
import re
def StopStem(df,stop_list):
#     stop = common_terms
    
    df['Lyrics'] = df['Lyrics'].apply(lambda x: re.sub(r'\[.*?\]',' ',x))
    
    df['Lyrics'] = df['Lyrics'].apply(lambda x: re.sub('[’‘!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]+',' ',x))
#     df["Lyrics"] = re.sub(r'[.*?]', '', df['Lyrics'].string)
    df["Lyrics"] = df['Lyrics'].str.replace('[^\w\s]',' ')
    df["Lyrics"] = df['Lyrics'].str.replace('作词',' ')
    df["Lyrics"] = df['Lyrics'].str.replace('作曲',' ')
    df["Lyrics"] = df['Lyrics'].str.replace('制作人',' ')
    df["Lyrics"] = df['Lyrics'].str.replace('编曲',' ')
    df["Lyrics"] = df['Lyrics'].str.replace('和声编写',' ')
    df["Lyrics"] = df['Lyrics'].str.replace('手鼓',' ')
    df["Lyrics"] = df['Lyrics'].str.replace('\r\n',' ')
    df['Lyrics'] = df['Lyrics'].apply(lambda x: ' '.join([word for word in jieba.cut(x) if word not in (stop_list)]))
#     join(list(jieba.cut(fileTrainRead[i][9:-11],cut_all=False)))
    return df

def cleanText(corpus,stop_list):
    stop_list = stop_list+['收起']
    preprocessed_sentences = []
#     stop_words = common_terms
#     pos_to_keep=('NOUN', 'PROPN', 'ADJ', 'VERB')
#     for doc in nlp.pipe(corpus, n_threads=8):
#         keep_tokens_string = ' '.join([t.lemma_ for t in doc if t.pos_ in pos_to_keep and t.lemma_ not in stop_words])
#         doc = nlp(lyrics)
#         lyrics = ' '.join([t.lemma_ for t in doc if t.pos_ in pos_to_keep and t.lemma_ not in stop_words])
#         preprocessed_sentences.append(keep_tokens_string)
    for text in corpus:
        text = [word for word in jieba.cut(text) if word not in (stop_list)]
        while ' ' in text:
            text.remove(' ')
        preprocessed_sentences.append(text)
#     preprocessed_sentences = 
    return preprocessed_sentences
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
n_dim = 500 

imdb_w2v = Word2Vec(size = n_dim, sample = 0.00001,)
imdb_w2v.build_vocab(X_train)
```

```python
imdb_w2v.train(X_train,total_examples = len(X_train),epochs = imdb_w2v.epochs)
```

```python
def buildWordVector(text, size):
    vec = np.zeros(size).reshape((1,size))
    count = 0.
    for word in text:
        try:
            vec += imdb_w2v[word].reshape((1,size))
            count += 1.
        except KeyError:
            continue
    if count != 0 :
        vec /= count
    return vec
```

```python
from sklearn.preprocessing import scale

train_vecs = np.concatenate([buildWordVector(z,n_dim) for z in X_train])
train_vecs = scale(train_vecs)
```

```python

imdb_w2v.train(X_test, total_examples = len(X_test), epochs = imdb_w2v.epochs)
test_vecs = np.concatenate([buildWordVector(z,n_dim) for z in X_test])
test_vecs = scale(test_vecs)
```

```python
from sklearn.linear_model import SGDClassifier

lr = SGDClassifier(loss = 'log', penalty='l1')
lr.fit(train_vecs, y_train)
print('test accuraccy: %.2f'%lr.score(test_vecs,y_test))
```

```python
lr = SGDClassifier(loss = 'log', penalty='l2')
lr.fit(train_vecs, y_train)
print('test accuraccy: %.2f'%lr.score(test_vecs,y_test))
```
