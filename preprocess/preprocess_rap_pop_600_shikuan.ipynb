```python
import nltk
from nltk.corpus import stopwords
```

```python
def StopStem(df):
    stop = set(stopwords.words('english'))
    df['Lyrics'] = df['Lyrics'].apply(lambda x: re.sub(r'\[.*?\]',' ',x))
    df['Lyrics'] = df['Lyrics'].apply(lambda x: re.sub('[’‘!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]+',' ',x))
    df["Lyrics"] = df['Lyrics'].str.lower().replace('[^\w\s]',' ')
    df["Lyrics"] = df['Lyrics'].str.replace('\r\n',' ')
    df['Lyrics'] = df['Lyrics'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
    return df
```

```python
pop_lyrics = StopStem(pop_lyrics)
```

```python
import ftfy
def fixUni(df):
    df['Lyrics'] = df['Lyrics'].apply(lambda x: ftfy.fix_text_encoding(x))
    return df
```

```python
songs = fixUni(df_general)
songs
# after fix unicode
```

```python
def remove_brackets(text):
    new_text = re.sub(r'\(.*?\)','',text)
    return new_text
```

```python
def ReBr(df):
        df['Lyrics'] = df['Lyrics'].apply(remove_brackets)
        return df
```

```python
def Chorus(df):
    words = ['Chorus','Verse1','Verse2','Verse']
    df['Lyrics'] = df['Lyrics'].apply(lambda x: ' '.join([word for word in x.split() if word not in words]))
    return df
```
