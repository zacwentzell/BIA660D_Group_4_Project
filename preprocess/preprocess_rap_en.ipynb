```python
import nltk
from nltk.corpus import stopwords
```

```python
def StopStem(df):
    stop = set(stopwords.words('english'))
    df['Lyrics'] = df['Lyrics'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
    df["Lyrics"] = df['Lyrics'].str.replace('[^\w\s]','')
    return df
```

```python
songs = StopStem(rap_lyrics)
```

```python
songs
# after remove stopwords and stem text
```

```python
import ftfy
def fixUni(df):
    df['Lyrics'] = df['Lyrics'].apply(lambda x: ftfy.fix_text_encoding(x))
    return df
```

```python
songs = fixUni(songs)
songs
# after fix unicode
```

```python
import re
def remove_brackets(text):
    new_text = re.sub(r'\(.*?\)','',text)
    new_text = re.sub(r'\[.*?\]','',text)
    return new_text
```

```python
import re
def ReBr(df):
        df['Lyrics'] = df['Lyrics'].apply(remove_brackets)
        return df
```

```python
def Chorus(df):
    words = ['Chorus','Verse1','Verse2','Verse']
    df['Lyrics'] = df['Lyrics'].apply(lambda x: ' '.join([word for word in x.split() if word not in words]))
    return df
```

```python
songs = ReBr(songs)
# after remove brackets
```

```python
songs = Chorus(songs)
# after remove unrelate words
```
